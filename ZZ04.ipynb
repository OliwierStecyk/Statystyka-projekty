### Zadanie 4: Korelacja liniowa [7 pkt]

#### a) Manualne obliczanie korelacji liniowej
Korzystając ze wzorów manualnie oblicz współczynnik korelacji liniowej `r` parametrów OD280/OD315 (oś X) i Flavanoids (oś Y).

```python

import pandas as pd
import numpy as np

# Wczytanie danych
data = pd.read_csv("wine_data.csv", delimiter=",", index_col=0)

# Wyświetlenie kilku pierwszych wierszy danych
print(data.head(10))

# Wybór kolumn 'OD280/OD315' i 'Flavanoids' z danych
OD280_OD315 = data['OD280/OD315']
Flavanoids = data['Flavanoids']

# Obliczenie średnich wartości
mean_OD280_OD315 = np.mean(OD280_OD315)
mean_Flavanoids = np.mean(Flavanoids)

# Obliczenie różnic między wartościami a ich średnimi
differences_OD280_OD315 = OD280_OD315 - mean_OD280_OD315
differences_Flavanoids = Flavanoids - mean_Flavanoids

# Obliczenie sumy iloczynów różnic
sum_differences_products = np.sum(differences_OD280_OD315 * differences_Flavanoids)

# Obliczenie sumy kwadratów różnic
sum_squared_differences_OD280_OD315 = np.sum(differences_OD280_OD315 ** 2)
sum_squared_differences_Flavanoids = np.sum(differences_Flavanoids ** 2)

# Obliczenie współczynnika korelacji
r = sum_differences_products / np.sqrt(sum_squared_differences_OD280_OD315 * sum_squared_differences_Flavanoids)

print("Współczynnik korelacji liniowej r:", r)
```

```python
# Komórka testowa
assert np.isclose(r, 0.7871939018669515)
```

### Zadanie 5: Regresja liniowa [10 pkt]

#### a) Manualne obliczanie współczynników równania regresji liniowej

Korzystając z metody najmniejszych kwadratów manualnie oblicz współczynniki równania regresji liniowej `a_1` i `b_1` parametru Flavanoids (oś Y) od OD280/OD315 (oś X).

```python

# Wczytanie danych
data = pd.read_csv("wine_data.csv", delimiter=",", index_col=0)

# Wybór odpowiednich kolumn jako dane X i Y
X = data['OD280/OD315']
Y = data['Flavanoids']

# Obliczenie współczynników regresji liniowej
mean_X = X.mean()
mean_Y = Y.mean()

diff_X = X - mean_X
diff_Y = Y - mean_Y

product_diff = diff_X * diff_Y
sum_product_diff = product_diff.sum()

sum_diff_X_squared = (diff_X ** 2).sum()

a_1 = sum_product_diff / sum_diff_X_squared
b_1 = mean_Y - a_1 * mean_X

print("Współczynnik a_1:", a_1)
print("Współczynnik b_1:", b_1)
```

```python
# Komórka testowa
assert np.isclose(a_1, 1.107473331774143)
assert np.isclose(b_1, -0.8631022610964245)
```

#### b) Obliczanie współczynników równania regresji liniowej z wykorzystaniem dedykowanych funkcji

Korzystając z funkcji `scipy.stats.linregress()` oblicz współczynniki równania regresji liniowej `a_2` i `b_2` parametru Flavanoids (oś Y) od OD280/OD315 (oś X).

```python

## Obliczenie współczynników regresji liniowej
result = sp.stats.linregress(data['OD280/OD315'], data['Flavanoids'])
a_2 = result.slope
b_2 = result.intercept
```

```python
# Komórka testowa
assert np.isclose(a_2, 1.1074733317741432)
assert np.isclose(b_2, -0.863102261096425)
```
